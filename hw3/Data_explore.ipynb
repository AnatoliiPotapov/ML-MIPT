### ,  item_id     .

```python
bins_count = 100
ranges = []

count, bins = np.histogram(train.item_id, bins = bins_count)
fl = False
for i in range(bins_count):
    if count[i]>0:
        if fl == False:
            print "Start: {0}".format(round(bins[i]))
            ranges.append(round(bins[i]))
        fl = True
    else:
        #if fl == True:
            #print "End: {0}".format(bins[i])
        fl = False

print " {0} ".format(len(ranges)+1)

def cluster_id(id):
    l = len(ranges)
    if id < ranges[0]:
        b_num = 0
    else:
        for i in range(l):
            if id > ranges[i]:
                b_num = i+1
    
    return b_num
        
map(cluster_id, [20440000, 20440000, 20440000])   
ranges
```

```python
def one_hot(column, values, first = 1):
    l = len(column)
    m = np.zeros((l, values))
    for i in range(l):
        m[i,column[i]- first] = 1
    return m
```

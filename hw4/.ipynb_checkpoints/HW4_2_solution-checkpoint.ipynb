# Выделить основу и суффиксы, которые меняются.

```python
def compare(w1,w2):
    l = min(len(w1), len(w2))
    res = []
    for i in range(l):
        if w1[i] != w2[i]:
            break
        res.append(w1[i])
            
    return res

def get_forms(w1, w2):
    l = len(compare(w1,w2))
    return [w1[:l], w1[l:], w2[l:]]
```

```python
def suff_len(suff):
    if suff == 'none':
        return 0
    else:
        return len(suff)
    
def get_ending(start_word, end_word, l):
    pointer = len(start_word) - l
    return end_word[pointer:]
    

def optimize_suff(iter = 1):
    
    for i in range(iter):
        results = []
        # для суффикса смотрим если его удлинение содержится в популярных стартовых суффиксах
        # и его новая замена есть в популярных суффиксах, то меняем
        start_pop = dict(train.start.value_counts())
        end_pop = dict(train.end.value_counts()[:10])       
        words = list(train['X'])
        end_word = list(train['form'])
        suffs = list(train['start'])
        
        for j in range(len(train)):
            for c in range(0,2):
                word = words[j]
                suff = suffs[j]
                new_suff = word[len(word)-len(suff)-1-c:]
                if not new_suff in start_pop:
                    pass
                else:
                    if start_pop[new_suff] > start_pop[suff]:
                        results.append((suff, new_suff))
                        train.set_value(j, 'start', new_suff)
                        train.set_value(j, 'end', get_ending(word, end_word[j], len(new_suff)))
                        
        
        print len(results)
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest
from sklearn.feature_extraction.text import TfidfVectorizer 
```

## Предиктор части речи

```python
class Predict_form_len(object):
    
    def __init__(self, regression = 'logistic'):
        self.seed = 42
        # create feature_union
        features = []
        features.append(('vectorizer',
            TfidfVectorizer(
                min_df=5,
                ngram_range=(2, 10),
                stop_words=None,
                decode_error='replace',
                analyzer='char',
                norm='l2',
                binary=False,
                max_features=150000
            ))
        )
        feature_union = FeatureUnion(features)
    
        #create estimator
        estimators = []
        estimators.append(('feature_union', feature_union))
        
        if regression == 'logistic':
            estimators.append(('regression', LogisticRegression(multi_class ='multinomial', solver ='newton-cg', penalty='l2', C = 1.0)))
        if regression == 'SVM':
            estimators.append(('regression', LinearSVC(penalty='l2')))
            
        self.model = Pipeline(estimators)
        
    def evaluate(self, X, y):
        # evaluate model
        kfold = KFold(n_splits=3, random_state=self.seed)
        results = cross_val_score(self.model, X, y, cv=kfold)
        print results.mean()
```

```python
form_predictor = Predict_form_len(regression = 'SVM')
form_predictor.evaluate(list(train.X), list(train.type))
```

```python
form_predictor.model.fit(train.X, train.type)
```

```python
accuracy_score(form_predictor.model.predict(train.X), train.type)
```

## Предиктор длины суффикса

```python
len_predictor = Predict_form_len(regression = 'SVM')
len_predictor.model.fit(train.X, train.s_len)
accuracy_score(len_predictor.model.predict(train.X), train.s_len)
```

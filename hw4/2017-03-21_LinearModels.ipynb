```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
X = np.sort(np.linspace(0, 1, 15))
y = X + 1 + np.random.normal(scale=1e-1, size=15)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

p = PolynomialFeatures(degree=10)
X_polynomial = p.fit_transform(X.reshape(-1, 1))
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_polynomial, y)
```

```python
X_test = np.linspace(1e-2, 1, 1000)
plt.plot(X_test, lr.predict(p.transform(X_test.reshape(-1, 1))))
plt.scatter(X, y)
plt.ylim(0, 3)
print lr.coef_
```

```python
lr.predict(p.transform(np.array(0).reshape(-1, 1)))
```

```python
from sklearn.linear_model import Ridge, Lasso

lr_regularized = Ridge(alpha=1e-2)
lr_regularized.fit(X_polynomial, y)
plt.plot(X_test, lr_regularized.predict(p.transform(X_test.reshape(-1, 1))))
plt.scatter(X, y)
print lr_regularized.coef_
```
